```python
"""
This source code is created by Pulung Hendro Prastyo
February,24,2020

Universitas Gadjah Mada, 
Faculty of Engineering, 
Department of Electrical and Information Technology Engineering
"""

#Import packages
import numpy as np
import cv2
from matplotlib import pyplot as plt
#take an image from library
image='image/img_1.png'
#convert color to grayscale
img = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
real_image = cv2.imread(image)
# default color of python is BGR, so we have to convert to RGB
real_image = cv2.cvtColor(real_image, cv2.COLOR_BGR2RGB)


#function of  median filter
def median_filter(img):
    
    #Median Filter
    img_out_median = img.copy()

    height = img.shape[0]
    width = img.shape[1]

    for i in np.arange(3, height-3):
        for j in np.arange(3, width-3):
            neighbors = []
            for k in np.arange(-3, 4):
                for l in np.arange(-3, 4):
                    a = img.item(i+k, j+l)
                    neighbors.append(a)
            neighbors.sort()
            median = neighbors[24]
            b = median
            img_out_median.itemset((i,j), b)
    return img_out_median

#function of min filter
def min_filter(img):
    #Min Filter
    img1 = img
    img_out_min = img1.copy()

    height = img1.shape[0]
    width = img1.shape[1]

    for i in np.arange(3, height-3):
        for j in np.arange(3, width-3):
            min = 255
            for k in np.arange(-3, 4):
                for l in np.arange(-3, 4):
                    a = img1.item(i+k, j+l)
                    if a < min:
                        min = a
            b = min
            img_out_min.itemset((i,j), b)
    return img_out_min


#call median filter
median_f = median_filter(img) 

#call min filter
min_f = min_filter(img)

#call min filter using a median image
median_min_f = min_filter(median_f)


#show images
plt.figure(figsize=(11,6))
plt.subplot(131)
plt.imshow(real_image,cmap='gray')
plt.title('Original Image')

plt.subplot(132)
plt.imshow(img,cmap='gray')
plt.title('Grayscale Image')

plt.subplot(133)
plt.imshow(median_f,cmap='gray')
plt.title('Median Filter')
plt.show()

plt.figure(figsize=(11,6))
plt.subplot(131)
plt.imshow(min_f,cmap='gray')
plt.title('Min Filter')

plt.subplot(132)
plt.imshow(median_min_f,cmap='gray')
plt.title('Median + Min Filter (My Purpose)')
plt.show()

```
