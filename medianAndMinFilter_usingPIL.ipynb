```python
"""
This source code is created by Pulung Hendro Prastyo
February,24,2020

Universitas Gadjah Mada, 
Faculty of Engineering, 
Department of Electrical and Information Technology Engineering
"""

# Importing Image and ImageFilter module from PIL package  
from PIL import Image, ImageFilter 
from matplotlib import pyplot as plt

# creating a image object 
image = Image.open("image/img_2.png",'r')
image1 = image.convert('L') #makes it greyscale

#im1 = cv2.imread('image/img_1.png')
#image2 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)

# applying the median filter 
median_filter = image1.filter(ImageFilter.MedianFilter(size = 5)) 
# applying the min filter 
min_filter = image1.filter(ImageFilter.MinFilter(size = 5))

# applying the min filter using median image
MedianMin_filter = median_filter.filter(ImageFilter.MinFilter(size = 5))

#Show images
plt.figure(figsize=(11,6))
plt.subplot(131)
plt.imshow(image,)
plt.title('Original Image')

plt.subplot(132)
plt.imshow(image1,cmap='gray')
plt.title('Grayscale Image')

plt.subplot(133)
plt.imshow(median_filter,cmap='gray')
plt.title('Median Filter')
plt.show()

plt.figure(figsize=(11,6))
plt.subplot(131)
plt.imshow(min_filter,cmap='gray')
plt.title('Min Filter')

plt.subplot(132)
plt.imshow(MedianMin_filter,cmap='gray')
plt.title('Median + Min Filter (My Purpose)')

plt.show()
```
